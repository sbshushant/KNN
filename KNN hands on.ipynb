```python
# Minkowski distance
# p and q, D dimensional space
# D(p,q) = (bar(x1-y1)p+bar(x2-y2)p +...+ bar(xn-yn)p)/(1/p)
# Manhattan distance = P = 1
# Euclidean distance = P = 2
# bar = mean
```

```python
# Manhattan distance = (x1-y1)+(x2-y2)+...+(xn-yn)
# It measures the distance between two points considering only horizontal and vertical movements

# Eucledian distance = sqrt(x1-y1)+ sqrt(x2-y2)+...+sqrt(xn-yn)
# It represents the shortest path between two points
```

```python
# replaceing null values
# if the dtype is int/float--->mean/median
# if the dtype is object --->mode
```

```python
col = list(df.columns)
```

```python
col
```

```python
for i in col:
    if df[i].dtype=="object":
        df[i] = df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```
